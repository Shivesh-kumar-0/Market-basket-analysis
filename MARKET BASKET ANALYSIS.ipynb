```python
import numpy as np
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=pd.read_csv('supermarket.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
fdata=data["MILK,BREAD,BISCUIT"].apply(lambda t: t.split(","))
```

```python
fdata=list(fdata)
fdata
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
fte=te.fit(fdata).transform(fdata)
```

```python
df=pd.DataFrame(fte,columns=te.columns_)
```

```python
df
```

```python
frequent_item=apriori(df,min_support=0.2,use_colnames=True)
```

```python
frequent_item
```

# association rule minning

```python
arm=association_rules(frequent_item,metric="confidence",min_threshold=0.3)
```
